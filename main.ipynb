```python
from collections import defaultdict,Counter
conldict=defaultdict(list)
for c in conl:
    for w in c:
        if w['head']==0:
            # root
            pass
        else:
            pw=c[w['head']-1]
            rel=w['deprel']
            w1=merge_dwp(w)
            w2=merge_dwp(pw)
            conldict[(w1,w2)].append(rel)
            conldict[(w2,w1)].append(rel)
for w in conldict:
    conldict[w]=Counter(conldict[w])
```

## corp

```python
corp=json.load(open('corpus.json','r',encoding="utf-8"))
```

```python
uposs=set()
for c in corp:
    for w in c:
        uposs.add(w['upos'])
print(len(uposs),uposs)
```

```python
def get_stats(puse=['ADJ','NOUN']):
    unidict={}
    bindict={}
    for c in corp:
        for i,w1 in enumerate(c):
            if is_puse(w1,puse):
                w=merge_dwp(w1)
                unidict[w]=unidict.get(w,0)+1
        for i in range(len(c)-1):
            if is_puse(c[i],puse) and is_puse(c[i+1],puse):
                w1=merge_dwp(c[i])
                w2=merge_dwp(c[i+1])
                bindict[(w1,w2)]=bindict.get((w1,w2),0)+1
    return unidict,bindict
```

# frequency

### only adj and noun

```python
unidict,bindict=get_stats(puse=['ADJ','NOUN'])
d2=sorted(bindict.items(), key=lambda d: d[1],reverse=True)
freq10=d2[:10]
to_pd_rel(freq10,conldict)
```

### all with punc

```python
unidict,bindict=get_stats(puse='all')
d2=sorted(bindict.items(), key=lambda d: d[1],reverse=True)
freq10=d2[:10]
to_pd_rel(freq10,conldict)
```

### all without punc

```python
unidict,bindict=get_stats(puse=list(uposs-set(['PUNCT'])))
d2=sorted(bindict.items(), key=lambda d: d[1],reverse=True)
freq10=d2[:10]
to_pd_rel(freq10,conldict)
```

# log likelihood ratio

```python
import copy
def lasmooth(unidict,bindict,la=1):
    ud=copy.deepcopy(unidict)
    bd=copy.deepcopy(bindict)
    V=len(unidict)
    for w in ud:
        ud[w]+=la*V
    for w in bd:
        bd[w]+=la
    return ud,bd

def Llog(k,n,x):
    return k*np.log(x)+(n-k)*np.log(1-x)

def lascore(c1,c2,c12,N):
    p=c2/N
    p1=c12/c1
    p2=(c2-c12)/(N-c1)
    l1=Llog(c12,c1,p)
    l2=Llog(c2-c12,N-c1,p)
    l3=Llog(c12,c1,p1)
    l4=Llog(c2-c12,N-c1,p2)
    lalog=l1+l2-l3-l4
    las=-2*lalog
    return las
def check(w1,w2):
    print(bindict.get((w1,w2)),unidict.get(w1),unidict.get(w2))
```

### only adj and noun

```python
unidict,bindict=get_stats(puse=['ADJ','NOUN'])
unidict,bindict=lasmooth(unidict,bindict,1e-10)
N=np.sum(list(unidict.values()))
ladict={}
for (w1,w2),c12 in bindict.items():
    c1=unidict.get(w1)
    c2=unidict.get(w2)
    if c1==0 or c2==0 :
        continue
    ladict[(w1,w2)]=lascore(c1,c2,c12,N)
d3=sorted(ladict.items(), key=lambda d: d[1],reverse=True)
la10=d3[:10]
to_pd_rel(la10,conldict)
```
